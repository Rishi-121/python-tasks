### _Task 1_

```python
import json

Associations = []

class general_query_log:
    
    # Read from the file "task1_json.txt"
    
    with open("./task1_json.txt", 'r') as f:
        
        __data = f.read()
    
    __data = json.loads(__data) # To parse the JSON string
    
    def __init__(self, WorkItemUId):
        
        self.WorkItemUId = WorkItemUId
        
        # Append the WorkItemUId inside the constructor
        
        Associations.append({"WorkItemUId": self.WorkItemUId})

    def fetch_data_on_parameters(self, key, value):
        
        self.key, self.value = key, value
        
        try:
            for i in range(len(general_query_log.__data)):
                
                for j in general_query_log.__data[i]["WorkItemAssociations"]:
                
                    if j[key] == value and j["WorkItemUId"] == self.WorkItemUId:
                        
                        Associations.append(j)
    
        except KeyError:
            
            print("Invalid Key")
```

```python
obj = general_query_log("528fbb5d-10a0-4c94-a1c0-c088c245d9d9")

# WorkItemTypeUId = '00020040-0200-0010-0110-000000000000' 

obj.fetch_data_on_parameters("WorkItemTypeUId", "00020040-0200-0010-0110-000000000000")

# WorkItemTypeUId = '00020040-0200-0010-0100-000000000000'

obj.fetch_data_on_parameters("WorkItemTypeUId", "00020040-0200-0010-0100-000000000000")

# AssociationTypeUId = '00200880-0040-0000-0000-000000000000'

obj.fetch_data_on_parameters("AssociationTypeUId", "00200880-0040-0000-0000-000000000000")
```
